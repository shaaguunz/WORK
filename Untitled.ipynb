```python
#since this is a tweet we have to remove the tags as an stopword
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
df['sentiment'].unique()
```

```python
df['sentiment']=df['sentiment'].replace("neutral","Neutral")
df['sentiment']=df['sentiment'].replace("positive","Positive")
df['sentiment']=df['sentiment'].replace("negative","Negative")
df['sentiment']=df['sentiment'].replace("Irrelevant",np.nan)
```

```python
df.dropna(inplace=True)
```

```python
names=['delta', 'deltaair', 'united', 'unitedair', 'southwest', 'southwestair', 'usairways',
         'virginamerica', 'american', 'americanair', 'jetblue', 'jetblues', 'usairway',
         'flight', 'airline', 'airlines','rt','said']
```

```python
def clean_data(x):
    text=x
    
    text = text.lower()
    text = re.sub('\[.*?\]', '', text) # remove square brackets
    text = re.sub(r'[^\w\s]','',text) # remove punctuation
    text = re.sub('\w*\d\w*', '', text) # remove words containing numbers
    text = re.sub('\n', '', text)
    return text
def clean_stopword(d):
    stop_words = stopwords.words('english')
    for name in names:
        stop_words.append(name)
    return " ".join([w.lower() for w in d.split() if w.lower() not in stop_words and len(w) > 1])

def tokenize(d):
    return word_tokenize(d)
```

```python
df['final_text']=df.text.apply(clean_data).apply(clean_stopword).apply(tokenize)
```

```python
df.drop('text',axis=1,inplace=True)
```

```python
df_fin=df.reset_index(drop=True)
```

```python
df_fin
```

```python
df_fin['final_text'] = df_fin.final_text.apply(lambda x: ', '.join([str(i) for i in x]))
```

```python
df_fin
```

```python
def sentiments(sentimento):
    if sentimento == "Positive":
        return 2
    elif sentimento == "Neutral":
        return 1
    elif sentimento == "Negative":
        return 0

# Applying the feeling function
df_fin.sentiment = df_fin.sentiment.apply(lambda x: sentiments(x))
print("Sentimentos:", df_fin.sentiment)
```

```python
X=df_fin.final_text
y=df_fin.sentiment
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB 
from sklearn.metrics import accuracy_score
import joblib
```
